```python
import numpy as np
from numpy.random import randn
```

```python
randn()
```

```python
x = randn()
if x > 1:
    answer = 'Greater than 1'
print(x)
print(answer)
```

```python
# After obtaining a number greater than 1; when we do it again it retains the value. 
# So we have to initialize it to None (it is same as Null in other languages)
```

```python
answer = None
x = randn()
if x > 1: 
    answer = 'Greater than 1'
print(x)
print(answer)
```

```python
answer = None
x = randn()
if x > 1: 
    answer = 'Greater than 1'
else:
    answer = 'Less than 1'
        
print(x)
print(answer)
```

```python
# Nested statements : 
answer = None
x = randn()
if x > 1: 
    answer = 'Greater than 1'
else:
    if x >= -1:
        answer = 'Between -1 and 1'
    else: 
        answer = 'Less than -1'
        
print(x)
print(answer)
```

```python
# Chained statements : 
answer = None
x = randn()
if x > 1: 
    answer = 'Greater than 1'
elif x >= -1:
    answer = 'Between -1 and 1'
else: 
    answer = 'Less than -1'
        
print(x)
print(answer)
```
